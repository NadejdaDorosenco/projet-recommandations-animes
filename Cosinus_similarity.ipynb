```python
import pandas as pd

# Charger les données depuis le fichier CSV
file_path = 'data/animes_update_base.csv'
anime_data = pd.read_csv(file_path)

# Afficher les premières lignes pour inspecter les données
anime_data.head()
```

```python
# Sélection et Prétraitement des Caractéristiques 
# Sélectionner les colonnes pertinentes pour la recommandation
selected_columns = ['title', 'synopsis', 'year', 'studios', 'genres', 'themes', 'source']
anime_features = anime_data[selected_columns]

# Afficher les premières lignes des données sélectionnées pour vérification
anime_features.head()

# Prétraitement des données
# Remplacer les NaN dans 'synopsis' et 'year' par des valeurs par défaut ou des marqueurs
anime_features['synopsis'] = anime_features['synopsis'].fillna('')
anime_features['year'] = anime_features['year'].fillna(anime_features['year'].median())

# Convertir les chaînes de caractères qui ressemblent à des listes en listes Python réelles
# Pour les colonnes 'studios', 'genres', et 'themes' qui sont sauvegardées comme des chaînes de caractères
import ast  # Pour convertir des chaînes en listes

def convert_string_to_list(string):
    try:
        # Convertir en liste si la chaîne est au format d'une liste
        return ast.literal_eval(string)
    except ValueError:
        # Sinon, retourner une liste vide ou une valeur par défaut
        return ['Unknown']

# Appliquer la conversion pour 'studios', 'genres', et 'themes'
anime_features['studios'] = anime_features['studios'].apply(convert_string_to_list)
anime_features['genres'] = anime_features['genres'].apply(convert_string_to_list)
anime_features['themes'] = anime_features['themes'].apply(convert_string_to_list)

# Vérifier les types de sources disponibles et uniformiser si nécessaire
anime_features['source'] = anime_features['source'].fillna('Unknown')

# Afficher les données après prétraitement pour vérification
anime_features.head()

```

```python
#Extraction de Caractéristiques avec TF-IDF
from sklearn.feature_extraction.text import TfidfVectorizer

# Initialiser le vecteur TF-IDF pour le synopsis
tfidf_vectorizer = TfidfVectorizer(stop_words='english', max_features=5000)

# Calculer le TF-IDF pour les synopsis
tfidf_matrix = tfidf_vectorizer.fit_transform(anime_features['synopsis'])

# Afficher la forme de la matrice TF-IDF obtenue pour vérification
tfidf_matrix.shape
```

```python
from sklearn.preprocessing import MultiLabelBinarizer

# Initialiser le binariseur pour les genres, thèmes et sources
genre_binarizer = MultiLabelBinarizer()
theme_binarizer = MultiLabelBinarizer()
source_binarizer = MultiLabelBinarizer()

# Encoder les genres, thèmes et sources
genres_binary = genre_binarizer.fit_transform(anime_features['genres'])
themes_binary = theme_binarizer.fit_transform(anime_features['themes'])
sources_binary = source_binarizer.fit_transform(anime_features['source'].apply(lambda x: [x]))

# Afficher la forme des matrices obtenues pour vérifier l'encodage
genres_binary.shape, themes_binary.shape, sources_binary.shape
```

```python
from scipy.sparse import hstack

# Concaténer les matrices TF-IDF et les matrices binaires pour former le vecteur de caractéristiques complet
feature_matrix = hstack([tfidf_matrix, genres_binary, themes_binary, sources_binary])

# Afficher la forme de la matrice des caractéristiques complète pour vérification
feature_matrix.shape
```

```python
#Définition de la Fonction de Recommandation
from sklearn.metrics.pairwise import cosine_similarity

import numpy as np

# Convertir la matrice de caractéristiques en format CSR pour l'indexation
feature_matrix_csr = feature_matrix.tocsr()

# Redéfinir la fonction de recommandation pour utiliser la matrice CSR
def recommend_animes_csr(anime_title, feature_matrix_csr, anime_features, top_n=10):
    # Trouver l'indice de l'anime donné
    idx = anime_features[anime_features['title'] == anime_title].index[0]
    
    # Calculer la similarité cosinus de cet anime avec tous les autres
    cosine_similarities = cosine_similarity(feature_matrix_csr[idx], feature_matrix_csr).flatten()
    
    # Obtenir les indices des animes les plus similaires
    similar_indices = np.argsort(-cosine_similarities)[1:top_n+1]
    
    # Retourner les titres des animes les plus similaires
    return anime_features['title'].iloc[similar_indices]

# Tester la fonction de recommandation 
recommended_animes_csr = recommend_animes_csr("Death Note", feature_matrix_csr, anime_features)
recommended_animes_csr



```

```python
# Modifier la fonction de recommandation pour exclure les animes contenant l'anime dans leur titre
def recommend_animes_exclude_keyword(anime_title, feature_matrix_csr, anime_features, exclude_keyword, top_n=10):
    # Trouver l'indice de l'anime donné
    idx = anime_features[anime_features['title'] == anime_title].index[0]
    
    # Calculer la similarité cosinus de cet anime avec tous les autres
    cosine_similarities = cosine_similarity(feature_matrix_csr[idx], feature_matrix_csr).flatten()
    
    # Obtenir les indices des animes les plus similaires
    similar_indices = np.argsort(-cosine_similarities)[1:]  # On prend tous les indices sauf le premier qui est l'anime lui-même
    
    # Filtrer pour exclure les titres contenant le mot-clé
    filtered_indices = [i for i in similar_indices if exclude_keyword.lower() not in anime_features['title'].iloc[i].lower()]
    
    # Retourner les titres des animes les plus similaires, sans ceux contenant le mot-clé
    return anime_features['title'].iloc[filtered_indices][:top_n]

# Appliquer la fonction modifiée pour obtenir des recommandations  sans inclure les animes liés
recommendations_without_Film = recommend_animes_exclude_keyword("Death Note", feature_matrix_csr, anime_features, "Death Note")
recommendations_without_Film
```
